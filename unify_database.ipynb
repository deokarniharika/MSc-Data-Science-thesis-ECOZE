```python
def standardize(df, mapping, store_name, missing_cols):
    df = df.rename(columns=mapping)
    for col in missing_cols:
        if col not in df:
            df[col] = None
    df['store'] = store_name
    return df

unified_columns = [
    "name", "brand", "price", "image", "description",
    "department", "aisle", "shelf", "recycling_info",
    "ingredients", "product_info", "store", "source_confidence"
]
```

```python
tesco_categories = [
    'Bakery', 'Food Cupboard', 'Drinks', 'Home & Ents', 'Fresh Food',
    'Baby & Toddler', 'Household', 'Health & Beauty', 'Treats & Snacks',
    'Frozen Food', 'Christmas', 'Pets'
]
```

```python
from itertools import chain
import pandas as pd

categories = {
    'waitrose': waitrose_dept,
    'sainsburys': sainsbury_dept,
    'lidl': lild_dept,
    'iceland': iceland_dept,
    'aldi': aldi_dept,
}

source_categories = list(set(filter(
    lambda x: isinstance(x, str), chain.from_iterable(categories.values())
)))
```

```python
from rapidfuzz import process, fuzz
```

```python
def map_to_tesco_categories(source_list, tesco_list, store_name, threshold=70):
    mapped = []
    for item in source_list:
        if not isinstance(item, str) or item.strip() == "":
            mapped.append((store_name, item, None, 0))
        else:
            match, score, _ = process.extractOne(item, tesco_list, scorer=fuzz.token_sort_ratio)
            mapped.append((store_name, item, match if score >= threshold else None, score))
    return mapped
```

```python
all_mappings = []

for store, dept_list in categories.items():
    mappings = map_to_tesco_categories(dept_list, tesco_categories, store)
    all_mappings.extend(mappings)

df_all_mappings = pd.DataFrame(all_mappings, columns=['Store', 'Original Department', 'Tesco Match', 'Score'])
```

```python
manual_fixes = {
    'Pet': 'Pets',
    'Beers, Wines & Spirits': 'Drinks',
    'Toiletries & Health': 'Health & Beauty',
    'Chilled Food': 'Fresh Food',
    'Vegetarian & Plant Based': 'Food Cupboard',
    # Add more as needed
}
df_all_mappings['Tesco Match'] = df_all_mappings.apply(
    lambda row: manual_fixes.get(row['Original Department'], row['Tesco Match']), axis=1
)
```

```python
from typing import List, Optional
import time
import re
import difflib

import vertexai
from google.cloud import aiplatform

PROJECT_ID = "uob-niharika-deokar"
REGION = "europe-west4"

aiplatform.init(project=PROJECT_ID, location=REGION)
vertexai.init(project=PROJECT_ID, location=REGION)

USE_GEMINI = True
try:
    from vertexai.generative_models import GenerativeModel
    gemini_model = GenerativeModel("gemini-2.5-flash-lite")
except Exception:
    USE_GEMINI = False
    from vertexai.language_models import ChatModel
    chat_model = ChatModel.from_pretrained("chat-bison@002")
    chat_session = chat_model.start_chat()

def build_prompt(dept: str, categories: List[str]) -> str:

    return (
        "You classify UK supermarket departments into ONE of the allowed Tesco categories.\n"
        "Rules:\n"
        " - Return EXACTLY one category from the allowed list (no extra words).\n"
        " - If uncertain, pick the closest plausible category.\n\n"
        f"Department to classify: {dept!r}\n"
        f"Allowed Tesco categories: {', '.join(categories)}\n"
        "Answer with only the category text from the allowed list."
    )

_ALLOWED_CACHE = None
def validate_or_snap(answer: str, allowed: List[str]) -> str:
    """
    Accept the answer if it's exactly in the allowed list;
    otherwise snap to the closest allowed label.
    """
    ans = (answer or "").strip()
    if ans in allowed:
        return ans

    ans = ans.strip(" '\"\n\t")
    if ans in allowed:
        return ans

    for c in allowed:
        if re.fullmatch(rf".*?\b{re.escape(c)}\b.*?", ans, flags=re.IGNORECASE):
            return c

    nearest = difflib.get_close_matches(ans, allowed, n=1, cutoff=0.0)
    return nearest[0] if nearest else allowed[0]

def classify_with_llm(dept: str, allowed: List[str]) -> Optional[str]:
    prompt = build_prompt(dept, allowed)
    try:
        if USE_GEMINI:
            resp = gemini_model.generate_content(prompt)
            text = getattr(resp, "text", None) or "".join(p.text for p in resp.candidates[0].content.parts)
        else:
            resp = chat_session.send_message(prompt, temperature=0)
            text = resp.text
        return validate_or_snap(text, allowed)
    except Exception as e:
        print(f"LLM error for '{dept}': {e}")
        return None

if 'unmapped_departments' not in globals():
    _unmapped_df = df_all_mappings[df_all_mappings['Tesco Match'].isna()]
    unmapped_departments = sorted([d for d in _unmapped_df['Original Department'].dropna().unique().tolist() if isinstance(d, str) and d.strip()])

print(f"Unmapped count before LLM: {len(unmapped_departments)}")


llm_mappings = {}
for dept in unmapped_departments:
    cat = classify_with_llm(dept, tesco_departments)
    if cat:
        llm_mappings[dept] = cat

    time.sleep(0.2)

print(f"LLM produced {len(llm_mappings)} mappings.")


def fill_llm(row):
    if pd.isna(row['Tesco Match']) and isinstance(row['Original Department'], str):
        return llm_mappings.get(row['Original Department'], row['Tesco Match'])
    return row['Tesco Match']

df_all_mappings['Tesco Match'] = df_all_mappings.apply(fill_llm, axis=1)


import pandas as pd
pd.DataFrame(
    [{"Original Department": k, "LLM Tesco Match": v} for k, v in llm_mappings.items()]
).to_csv("llm_department_mappings.csv", index=False)


remaining_unmapped = df_all_mappings['Tesco Match'].isna().sum()
print(f"Remaining unmapped after LLM: {remaining_unmapped}")
```
